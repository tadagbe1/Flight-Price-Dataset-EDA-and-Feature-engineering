# Starting the feature engineering process

## Handling the Date_of_Journey column

```python
df['Day'] = df['Date_of_Journey'].str.split('/').str[0] 
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df
```

```python
df.info()
```

```python
df['Day'] = df['Day'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df
```

```python
df.info()
```

```python
# Drop Date of Journey

df.drop('Date_of_Journey', inplace=True, axis=1)
df.head()
```

## Handling the Arrival_Time's column

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_minute'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df
```

```python
df['Arrival_hour'].astype(int)
df['Arrival_minute'].astype(int)

df.drop('Arrival_Time', axis=1, inplace=True)
df.head()
```

## Handling Dep_Time column

```python
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_minute'] = df['Dep_Time'].str.split(':').str[1]

df['Dep_hour'].astype(int)
df['Dep_minute'].astype(int)

df.drop('Dep_Time', axis=1, inplace=True)


df.head()
```

## Handling columns categorical data

### Handling Total_Stops column

```python
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, np.nan:1})
```

```python
df[df['Total_Stops'].isnull()]
```

### Handling the Route column

```python
df.drop('Route', axis=1, inplace=True)
```

```python
df.head()
```

### Handling Airline's, Source's and Destination's column using One Hot Encoding

```python
# We import the OneHotEncode from sklearn
from sklearn.preprocessing import OneHotEncoder
```

```python
#create an instance of OneHotEncoder, fit and transform
encoder=OneHotEncoder()
encoder_fit_tra=encoder.fit_transform(df[['Airline', 'Source', 'Destination']]).toarray()
```

```python
encoder_df = pd.DataFrame(encoder_fit_tra, columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
df = pd.concat([df, encoder_df], axis=1)

#df.drop(['Airline', 'Source', 'Destination'], inplace=True, axis=1)
```

```python
df.head()
```

```python
df.drop(['Airline', 'Source', 'Destination'], inplace=True, axis=1)
```

```python
df.head()
```

### Handling the Duration column

```python
df['Duration'] = df['Duration'].apply(lambda x : '0h ' +x if 'h' not in x else x+' 0m' if 'm' not in x else x)

df['Duration_hour'] = df['Duration'].str.split(' ').str[0]
df['Duration_minute'] = df['Duration'].str.split(' ').str[1]

df['Duration_hour'] = df['Duration_hour'].str.split('h').str[0]
df['Duration_minute'] = df['Duration_minute'].str.split('m').str[0]

```

```python


df['Duration_hour'] = df['Duration_hour'].fillna(0)
df['Duration_minute'] = df['Duration_minute'].fillna(0)
```

```python
df['Duration_hour'] = df['Duration_hour'].astype(int)
df['Duration_minute'] = df['Duration_minute'].astype(int)
```

```python
df.head()
```

```python
df.drop('Duration', inplace=True, axis=1)
df.head()
```
